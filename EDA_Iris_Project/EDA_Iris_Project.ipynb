# Iris Dataset EDA
Using NumPy, Pandas, Matplotlib and Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('iris.csv')
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
print("Duplicates:",df.duplicated().sum())
```

```python
num=df.select_dtypes(include='number')
for c in num.columns:
    plt.figure(figsize=(5,3))
    sns.histplot(df[c],kde=True)
    plt.title(c)
    plt.show()
```

```python
if 'species' in df.columns:
    plt.figure(figsize=(5,3))
    sns.countplot(data=df,x='species')
    plt.show()
```

```python
for c in num.columns:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[c])
    plt.title(c)
    plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(num.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
sns.pairplot(df,hue='species' if 'species' in df.columns else None)
plt.show()
```

## Conclusion
- Checked shape, datatypes, missing values and duplicates.
- Visualized distributions and outliers.
- Generated correlation heatmap and pairplot.
- Dataset is clean and suitable for ML.
